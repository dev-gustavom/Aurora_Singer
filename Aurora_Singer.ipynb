# **Análise Energética**

Considerando:

Capacidade total: 1000 kWh

Carga atual: 80%

Consumo na decolagem: 300 kWh

Perdas energéticas: 10%

```python
# Dados iniciais
capacidade_total = 1000  # kWh (capacidade máxima da bateria)
carga_atual_pct = 80     # porcentagem atual
consumo_decolagem = 300  # energia usada na decolagem (kWh)
perda_pct = 10           # perdas energéticas (%)

# Converter porcentagem em valor real
energia_disponivel = capacidade_total * (carga_atual_pct / 100)

# Calcular perdas
perdas = energia_disponivel * (perda_pct / 100)

# Energia útil após perdas
energia_util = energia_disponivel - perdas

# Energia restante após decolagem
energia_restante = energia_util - consumo_decolagem

# Mostrar resultados
print("Energia disponível:", energia_disponivel, "kWh")
print("Perdas energéticas:", perdas, "kWh")
print("Energia útil:", energia_util, "kWh")
print("Energia restante após decolagem:", energia_restante, "kWh")

# Verificar se há energia suficiente para decolagem

if energia_restante < 0:
    print("Energia insuficiente para decolagem!")
elif energia_restante < 200:
    print("Decolagem possível, mas com baixo nível de segurança!")
else:
    print("Energia suficiente para decolagem segura!")
```
